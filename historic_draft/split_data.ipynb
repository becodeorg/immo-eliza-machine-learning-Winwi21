```python
import numpy as np 
import pandas as pd 
```

```python
data = './data/Kangaroo.csv'
kangaroo = pd.read_csv(data, sep = ',')
df_kangaroo = pd.DataFrame(kangaroo)

print(df_kangaroo)
```

1. Split Dataset : Data_test & Data_train

```python
import random

# Choose 30% of dataset for test
df_30 = df_kangaroo.sample(frac=0.3, random_state=42)
print(df_30)
df_30.to_csv('data_fortest.csv', index=False)
```

```python
# Collect the rest data for training
df_70 = df_kangaroo.drop(df_30.index)
print(df_70)
df_70.to_csv('data_fortrain.csv', index=False)

```

```python
# Evaluate the missing values for df_30:
na_values = df_30.isnull().sum()
missing_percent = (na_values/ len(df_30)) * 100
print (missing_percent.round(2))
print(df_30.shape)
```

```python
# Evaluate the missing values for df_70:
na_values = df_70.isnull().sum()
missing_percent = (na_values/ len(df_70)) * 100
print (missing_percent.round(2))
print(df_70.shape)
```
