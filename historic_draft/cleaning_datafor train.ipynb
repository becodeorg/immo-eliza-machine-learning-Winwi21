DATA CLEANING

```python
import numpy as np 
import pandas as pd 
```

```python
data = './data_fortrain.csv'
data_train = pd.read_csv(data, sep = ',')
df_70 = pd.DataFrame(data_train)

print(df_70)
```

```python
# Search if there are duplicates and remove it
duplicates = df_70.duplicated()
df_70.drop_duplicates(inplace=True)
print(duplicates.count)
```

```python
na_values = df_70.isnull().sum()
missing_percent = (na_values/ len(df_70)) * 100
print (missing_percent.round(2))
```

```python
for col in df_70.columns:
    if missing_percent[col] > 50:
        df_70.drop(columns=col, inplace=True)
print(df_70.info)
```

```python
#Remove unuseful and unrelevant columns: 
df_70= df_70.drop(columns=['url','id','kitchenType','toiletCount','buildingConstructionYear','facedeCount', 'floodZoneType','locality','hasLivingRoom'], errors='ignore')
print(df_70)
```

```python
# Display head
df_70.head(10)
```

```python
# Delete row when the price == 0
df_70 = df_70[df_70['price'] != 0]
print(df_70)
```

```python
df_70.head(10)
```

```python
# Replace NaN to 0 : 
df_70 = df_70.fillna(0)

print(df_70)
```

```python
#Replace "True" by 1
df_70 = df_70.replace(True, 1)
print(df_70)
```

```python
df_70.head(10)
```

```python
df_70.to_csv("df70_clean.csv", index=False, encoding='utf-8')
```
